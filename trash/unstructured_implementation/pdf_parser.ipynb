```python
from py_pdf_parser.loaders import load_file

document = load_file("../data/pacific-rim-2013_1.pdf")
```

```python
document.pages[0].width
```

```python
def is_all_uppercase(input_string):
    # Check if all characters in the string are uppercase
    input_string=input_string.strip()
    return all(char.isupper() for char in input_string if char.isalpha())
```

```python
import re

def is_int_ext_line(line):
    line=line.strip()
    if line.startswith("INT.") or line.startswith("EXT."):
        return True
    else:
        return False
```

```python
scene_info_coordinates={}
for i in document.elements:
    text_temp=i.text()
    if is_int_ext_line(text_temp):
        scene_info_coordinates[text_temp]=i.original_element.bbox
```

```python
scene_info_coordinates
```

```python
from collections import Counter
value_counter=Counter(scene_info_coordinates).values()
max_count = max(value_counter)
```

```python
thresold_xmin=max_count[0]
thresold_xmax=max_count[2]
```

```python
movie_info_coordinates={}
for index,i in enumerate(document.elements):
    index=str(index)+":number:"
    text_temp=i.text()
    print(text_temp)
    cooridinates=i.original_element.bbox
    if is_int_ext_line(text_temp):
        movie_info_coordinates[index+text_temp]={"metadata":"scene_info","bbox":cooridinates}
    elif 0.95 * max_count[0] <= int(cooridinates[0]) <= 1.05 * max_count[0]:
        movie_info_coordinates[index+text_temp]={"metadata":"action_line","bbox":cooridinates}
    elif (is_all_uppercase(text_temp)) and (thresold_xmin < cooridinates[0]):
        movie_info_coordinates[index+text_temp]={"metadata":"character","bbox":cooridinates}
    elif (thresold_xmin < cooridinates[0]) and (thresold_xmax>cooridinates[-1]):
        movie_info_coordinates[index+text_temp]={"metadata":"dialogue","bbox":cooridinates}
    else:
        movie_info_coordinates[index+text_temp]={"metadata":"unknown","bbox":cooridinates}
        
```

```python
movie_info_coordinates
```

```python
import json
with open("json_file_path.json", 'w') as json_file:
    json.dump(movie_info_coordinates, json_file)
```
