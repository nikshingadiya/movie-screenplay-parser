## Replace Title With Scene

```python
import re

def is_int_ext_line(line):
    line=line.strip()
    if line.startswith("INT.") or line.startswith("EXT."):
        return True
    else:
        return False
    
```

```python
for index,(i,j) in enumerate(zip(text_list,text_type_list)):
    
    if is_int_ext_line(i) and j=="Title":
        text_type_list[index]="scene_info"
        print(i)
    
```

```python
with open("inter.json","w") as file:
    json.dump(x,file)
```

```python
import re

transitionEnum = [
    "FADE IN",
    "FADE OUT",
    "FADE UP",
    "JUMP CUT",
    "MATCH CUT",
    "SMASH MATCH CUT",
    "MATCH DISSOLVE",
    "CUT",
    "DISSOLVE",
    "FLASH CUT",
    "FREEZE FRAME",
    "IRIS IN",
    "IRIS OUT",
    "WIPE TO",
    "CUT TO"
]

def checkTransition(line):
    line = line.strip()
    transitions_pattern = re.compile(r'\b(?:' + '|'.join(re.escape(transition) for transition in transitionEnum) + r')\b', re.IGNORECASE)
    return bool(re.search(transitions_pattern, line))
```

## Transition Replacement

```python

for index,(i,j) in enumerate(zip(text_list,text_type_list)):
    
    if (checkTransition(i) and j=="UncategorizedText") or (checkTransition(i) and j=="Title"):
        text_type_list[index]="transition"
        print(i)
        
```

```python

def create_and_save_json(text_list, text_type_list, filename="inter.json"):
    x = dict(zip(text_list, text_type_list))
    with open(filename, "w") as file:
        json.dump(x, file)

create_and_save_json(text_list=text_list,text_type_list=text_type_list)
```

## Page Number Replacement

```python
import re

def contains_only_numbers(line):
    line=line.strip().replace(".","")
    return bool(re.match(r'^\d+$', line))
```

```python

for index,(i,j) in enumerate(zip(text_list,text_type_list)):
    
    if (contains_only_numbers(i) and j=="UncategorizedText"):
        text_type_list[index]="page_number"
        print(i)
create_and_save_json(text_list=text_list,text_type_list=text_type_list)
```
